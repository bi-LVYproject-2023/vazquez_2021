```python
from Bio import SeqIO
```

```python
fasta = SeqIO.parse('../data/genomes/annotations/mamColU_proteins.fasta', 'fasta')

ids = []
seqs = []

for seq in fasta:
    ids.append(seq.id)
    seqs.append(''.join(seq.seq))
    
prots = pd.DataFrame({'id': ids, 'seq': seqs})
```

```python
prots['len'] = prots.seq.apply(len)
```

```python
prots
```

```python
prots[prots['len'] > 100]
```

```python
prots[prots['len'] > 300]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_bar
```

```python
bar = sns.barplot(df_bar, x='Metric', y='Count')
plt.title('Main metrics for annotated genome of Mammuthus columbi')
bar.set(xlabel=None)
plt.savefig('../figures/blat_mammoth_stats.png')
```
